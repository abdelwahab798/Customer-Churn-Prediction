```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\nice\Desktop\m\data\clean Data.csv")
```

```python
df.head()
```

```python
fig , ax = plt.subplots(figsize=(20,15),nrows=4,ncols=3)
sns.countplot(x=df["gender"],hue=df["Churn"],ax=ax[0][0])
ax[0][0].set_title("Gender Type Count")
sns.countplot(x=df["SeniorCitizen"],hue=df["Churn"],ax=ax[0][1])
ax[0][1].set_title("SeniorCitizen Type Count")
sns.countplot(x=df["Partner"],hue=df["Churn"],ax=ax[0][2])
ax[0][2].set_title("Partner Type Count")
sns.countplot(x=df["Dependents"],hue=df["Churn"],ax=ax[1][0])
ax[1][0].set_title("Dependents Type Count")
sns.countplot(x=df["PhoneService"],hue=df["Churn"],ax=ax[1][1])
ax[1][1].set_title("PhoneService Type Count")
sns.countplot(x=df["MultipleLines"],hue=df["Churn"],ax=ax[1][2])
ax[1][2].set_title("MultipleLines Type Count")
sns.countplot(x=df["InternetService"],hue=df["Churn"],ax=ax[2][0])
ax[2][0].set_title("InternetService Type Count")
sns.countplot(x=df["OnlineSecurity"],hue=df["Churn"],ax=ax[2][1])
ax[2][1].set_title("OnlineSecurity Type Count")
sns.countplot(x=df["OnlineBackup"],hue=df["Churn"],ax=ax[2][2])
ax[2][2].set_title("OnlineBackup Type Count")
sns.countplot(x=df["DeviceProtection"],hue=df["Churn"],ax=ax[3][0])
ax[3][0].set_title("DeviceProtection Type Count")
sns.countplot(x=df["TechSupport"],hue=df["Churn"],ax=ax[3][1])
ax[3][1].set_title("TechSupport Type Count")
sns.countplot(x=df["StreamingTV"],hue=df["Churn"],ax=ax[3][2])
ax[3][2].set_title("StreamingTV Type Count")
plt.tight_layout()
plt.show()

```
